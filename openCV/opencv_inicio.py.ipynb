```python
img_zero=np.zeros(shape=(500,500,3),dtype=np.int16)
```

```python
plt.imshow(img_zero)
```

```python
linea_roja= cv2.line(img_zero,(0,0),(500,500),(255,0,0),10)
```

```python
cuadrado= cv2.rectangle(img_zero,(500,0),(400,100),(0,0,255),5)
plt.imshow(cuadrado)
```

```python
circulo=cv2.circle(img_zero,(300,64),63,(0,255,0),-1)
plt.imshow(circulo)
```

```python
##cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
font = cv2.FONT_HERSHEY_SIMPLEX
text=cv2.putText(img_zero,"Robin",(10,500),font,4,(255,255,255),2,cv2.LINE_4)
plt.imshow(text)
```
